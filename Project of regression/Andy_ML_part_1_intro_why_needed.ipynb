```python
import pandas as pd

df = pd.read_csv("house_price.csv")
display(df.head())
```

```python
x_data = np.array(df['size'])
y_data =np.array(df.price)
```

```python
slope, y_intercept = np.polyfit(x_data, y_data, 1)

print(f"Calculated Slope: {slope:.2f}")
print(f"Calculated Y-intercept: {y_intercept:.2f}")

# Optional: Visualize the scatterplot and the fitted line
plt.scatter(x_data, y_data, label='Data Points')
plt.plot(x_data, slope * x_data + y_intercept, color='red', label='Fitted Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatterplot with Linear Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
degree = 2

# Calculate the coefficients
coefficients = np.polyfit(x_data, y_data, degree)
print(f"Calculated Coefficients (Degree {degree}): {coefficients}")
```

```python

# Create a polynomial function from the coefficients
poly_function = np.poly1d(coefficients)

# Optional: Visualize the scatterplot and the fitted line
plt.scatter(x_data, y_data, label='Data Points')

# Create a smoother line for visualization
# Generate a range of x values for the fitted line
x_fit = np.linspace(min(x_data), max(x_data), 100)
plt.plot(x_fit, poly_function(x_fit), color='red', label=f'Fitted Line (Degree {degree})')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(f'Scatterplot with Polynomial Regression Line (Degree {degree})')
plt.legend()
plt.grid(True)
plt.show()
```
